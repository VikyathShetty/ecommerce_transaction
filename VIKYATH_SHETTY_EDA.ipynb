## Merging Dataset

```python
df_merged = df_transactions.merge(df_customer, on="CustomerID", how="left")
df_merged.head()
```

```python
df_merged = df_merged.merge(df_products, on='ProductID', how='left')
df_merged.head(50)
```

```python
print(df_merged["Price_y"].mean())

print(df_merged["Price_x"].mean())
```

```python
if 'Price_x' in df_merged.columns and 'Price_y' in df_merged.columns:
    if df_merged['Price_x'].equals(df_merged['Price_y']):
        df_merged.drop(columns=['Price_y'], inplace=True)
        df_merged.rename(columns={'Price_x': 'Price'}, inplace=True)
```

## Data visualization

## Monthly Sales Trend Analysis

```python
df_merged['Month'] = df_merged['TransactionDate'].dt.to_period("M")
monthly_sales = df_merged.groupby('Month')['TotalValue'].sum()

plt.figure(figsize=(12, 5))


sns.barplot(x=monthly_sales.index.astype(str), y=monthly_sales.values, color="grey", label="Total Sales")


sns.lineplot(x=monthly_sales.index.astype(str), y=monthly_sales.values, marker="*", color="black", label="Sales Trend")

plt.xticks(rotation=45)
plt.title("Monthly Sales Trend (Bar + Line Chart)")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

## Analysis of  High-Value Customers

```python
customer_spending = df_merged.groupby('CustomerID')['TotalValue'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=customer_spending.values, y=customer_spending.index)
plt.title("Top 10 High-Value Customers")
plt.xlabel("Total Spending")
plt.ylabel("Customer ID")
plt.show()
```

## Regional Sales & Customer Distribution

```python

plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)
region_sales = df_merged.groupby('Region')['TotalValue'].sum().sort_values(ascending=False)
sns.barplot(x=region_sales.index, y=region_sales.values, palette="magma")
plt.title("Total Sales Value by Customer Region")
plt.xlabel("Region")
plt.ylabel("Total Sales Value")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.subplot(1, 2, 2)
region_counts = df_customer['Region'].value_counts()
sns.barplot(x=region_counts.index, y=region_counts.values, palette="viridis")
plt.title("Customer Count by Region")
plt.xlabel("Region")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## Top Selling Products from each Category

```python
top_products_by_category = (
    df_merged.groupby(['Category', 'ProductName'])['Quantity']
    .sum()
    .reset_index()
    .sort_values(['Category', 'Quantity'], ascending=[True, False])
    .groupby('Category')
    .head(3)
)

plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_products_by_category, 
    x='Quantity', 
    y='ProductName', 
    hue='Category', 
    dodge=False, 
    palette='viridis'
)
plt.title("Top 3 Selling Products in Each Category")
plt.xlabel("Total Quantity Sold")
plt.ylabel("Product Name")
plt.legend(title="Category")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

## Insights on Customer Spending Based on Signup Duration

```python
today = pd.Timestamp.today()


df_merged['SignupCategory'] = df_merged['SignupDate'].apply(
    lambda x: "More than 1 Year" if (today - x).days > 365 else "Less than 1 Year"
)


signup_spending_mean = df_merged.groupby('SignupCategory')['TotalValue'].mean().reset_index()


print("Average Total Spending for each Signup Category:")
print(signup_spending_mean)


plt.figure(figsize=(8, 5))
ax = sns.barplot(data=signup_spending_mean, x='SignupCategory', y='TotalValue', palette='coolwarm')


for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

plt.title("Average Spending Based on Signup Duration")
plt.xlabel("Signup Duration Category")
plt.ylabel("Average Total Spending")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


signup_customer_count = df_merged.groupby('SignupCategory').size().reset_index(name='CustomerCount')

print("Total Number of Customers in Each Signup Category:")
print(signup_customer_count)

```
